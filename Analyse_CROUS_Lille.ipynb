```python
counts = df_crous_lille['type'].value_counts()
plt.figure()
colors=["#e37820","#225de6"]
plt.pie(counts.values, labels= counts.index, autopct = '%1.1f%%',startangle=90,colors=colors)
plt.title("Repartition des servivces CROUS a Lille")
plt.show()
```

```python
import matplotlib.pyplot as plt

counts_zone = df_crous_lille['zone'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    counts_zone.values,
    labels=counts_zone.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Répartition des services CROUS par zone")
plt.show()
```

```python
import matplotlib.pyplot as plt

counts_zone = df_crous_lille['zone'].value_counts()

plt.figure(figsize=(10,5))
counts_zone.plot(kind='bar')
plt.title("Nombre de services CROUS par zone")
plt.xlabel("Zone")
plt.ylabel("Nombre de services")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Séparer les restos et cafétérias
restos = df_crous_lille[df_crous_lille['type'] == 'resto']
cafets = df_crous_lille[df_crous_lille['type'] == 'cafeteria']

plt.figure(figsize=(8,6))

# Scatter plot pour les restos
plt.scatter(restos['lon'], restos['lat'], color='green', label='Resto', s=100, alpha=0.7)

# Scatter plot pour les cafétérias
plt.scatter(cafets['lon'], cafets['lat'], color='orange', label='Cafeteria', s=100, alpha=0.7)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Localisation des services CROUS à Lille")
plt.legend()
plt.grid(True)
plt.show()
```
