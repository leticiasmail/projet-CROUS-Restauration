```python
# Compter les occurrences par zone
counts_zone = df_crous_lille['zone'].value_counts()

# Si trop de zones, tu peux regrouper les petites en 'Autres'
top_zones = counts_zone[counts_zone > 1]  # garde les zones avec plus d'un resto
other = counts_zone[counts_zone <= 1].sum()  # somme des petites zones
top_zones['Autres'] = other

# Couleurs
colors_zone = plt.cm.tab20.colors  # palette de 20 couleurs

# Création de la pie chart
plt.figure(figsize=(8,8))
plt.pie(top_zones, labels=top_zones.index, autopct='%1.1f%%', startangle=90, colors=colors_zone)
plt.title("Répartition des restos CROUS par zone")
plt.show()
```
